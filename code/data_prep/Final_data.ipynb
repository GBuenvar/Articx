## Read the dataset

The dataset `Matrix_dataset_Final.csv` contains the information about the relationships between dependent and idependent variables. 

```python
dataset_final = pd.read_csv("../../data/Matrix_dataset_FINAL.csv")
dataset_final = dataset_final.iloc[1:,:]
dataset_final = dataset_final.dropna()
dataset_final
```

```python
unique_values = dataset_final["Main Score"].unique()
print("set of unique values in Main Score: ", unique_values)
```

```python

indep = dataset_final.Independent.unique()
n = len(indep)
print(n)
Matrix = np.zeros((n,n), dtype=int).astype(str)

for r, row in dataset_final.iterrows():
    Matrix[indep == row.Dependent, indep == row.Independent] = row["Main Score"]
Matrix
```
